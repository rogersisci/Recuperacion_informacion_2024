## **P6. ¿Qué piensas hacer con los datos erroneos?**

```python
datai["precipitacion"] = np.where(datai["precipitacion"] == -99.9, np.nan, datai["precipitacion"])
datap["precipitacion"] = np.where(datap["precipitacion"] == -99.9, np.nan, datap["precipitacion"])
datai.isnull().sum()
```

```python
datap.isnull().sum()
```

```python
datai["precipitacion"] = datai["precipitacion"].ffill()
datap["precipitacion"] = datap["precipitacion"].ffill()
datai.isnull().sum()
```

```python
datap.isnull().sum()
```

```python

datai["Time_steps"] = pd.to_datetime((datai.año*10000 + datai.mes*100 + datai.dia).apply('{:.0f}'.format),format='%Y%m%d')

datai["Time_steps"] = pd.to_datetime(datai['año'].astype(str)+' '+datai['mes'].astype(str) +' '+ datai['dia'].astype(str),format='%Y %m %d').dt.strftime('%Y-%m-%d')

datap["Time_steps"] = pd.to_datetime((datap.año*10000 + datap.mes*100 + datap.dia).apply('{:.0f}'.format),format='%Y%m%d')

datap["Time_steps"] = pd.to_datetime(datap['año'].astype(str)+' '+datap['mes'].astype(str) +' '+ datap['dia'].astype(str),format='%Y %m %d').dt.strftime('%Y-%m-%d')
datai.head()
```

```python
data_ni["Time_steps"]=pd.to_datetime(data_ni['YEAR'].astype(str)+' '+data_ni['DOY'].astype(str),format='%Y %j').dt.strftime('%Y-%m-%d')
data_ni.head()
```

```python
data_np["Time_steps"]=pd.to_datetime(data_np['YEAR'].astype(str)+' '+data_np['DOY'].astype(str),format='%Y %j').dt.strftime('%Y-%m-%d')
data_np.head()
```

```python
data_ni2=data_ni[(data_ni['Time_steps'] > '2015-11-01') & (data_ni['Time_steps'] < '2024-06-01')]
data_ni2[['Time_steps', 'precipitacion']]
```

```python
data_np2=data_np[(data_np['Time_steps'] > '2014-05-01') & (data_np['Time_steps'] < '2024-06-01')]
data_np2[['Time_steps', 'precipitacion']]
```

```python
frames_imata = [datai[['Time_steps', 'precipitacion']], data_ni2[['Time_steps', 'precipitacion']]]
result_imata = pd.concat(frames_imata)
result_imata
```

```python
frames_pillones = [datap[['Time_steps', 'precipitacion']], data_np2[['Time_steps', 'precipitacion']]]
result_pillones = pd.concat(frames_pillones)
result_pillones.to_excel("output.xlsx")
result_pillones
```
